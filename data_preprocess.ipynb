```python
import pandas as pd
```

```python
raw_data = pd.read_csv("data/eastmoney_topic.csv")
```

```python
raw_data.columns
```

```python
raw_data.describe(include='all')
```

select relevant columns

```python
data = raw_data[['item_title', 'item_author_cate', 'article_author', 'article_source_cate', 'month',
       'eastmoney_robo_journalism', 'media_robo_journalism', 'SMA_robo_journalism', 'viral', 'sentiment_score', 'topics']]
```

process topics information

```python
data['topics']
```

```python
data['topics']
```

```python
data['topics'] = data['topics'].apply(lambda x: x.strip('()').split('],')[0])
```

```python
data['topics_val'] = data['topics'].apply(lambda x: [float(tup.split(',')[1][1:-1]) for tup in x.strip('[]').split('),')])
```

```python
data['topics_val']
```

```python
# Convert strings to lists
# data['topics_val'] = data['topics_val'].apply(lambda x: list(map(float, x.strip('[]').split(','))))

# Create separate columns for each value
data[['topics_val1', 'topics_val2', 'topics_val3', 'topics_val4', 'topics_val5']] = pd.DataFrame(data['topics_val'].tolist())
```

```python
data[['topics_val1', 'topics_val2', 'topics_val3', 'topics_val4', 'topics_val5']]
```

```python
data = data.drop(columns=['topics', 'topics_val'])
```

include stock code

```python
stock_data = pd.read_csv('./data/eastmoney_bert.csv')
```

```python
data['stock_code'] = stock_data['stock_code']
```

output processed dataset

```python
data.to_csv('data/eastmoney_bert.csv', index=False)
```
